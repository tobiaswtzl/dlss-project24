# Create datasets for different analysis

```python
#Import necessary libraries
import pandas as pd

#Load raw and preprocessed data
raw_posts = pd.read_csv('data/raw/total_posts.csv')
posts = pd.read_csv('data/preprocessed/total_posts.csv')
```

```python
# Merge subreddit name and id on preprocessed data to filter them
sub_posts = posts.merge(raw_posts[['id', 'subreddit.id', 'subreddit.name']], on='id', how='left')

# Define the list of subreddits to process
subreddits = ['genz', 'millennials', 'askmen', 'askwomen', 'askscience', 'democrats', 'republican', 'politics', 'conspiracy']

# Emptly dict to store the length of each subreddit data
posts_lengths = {}

# Process sub_posts
for subreddit in subreddits:
    # Filter data for the current subreddit
    sub_posts_filtered = sub_posts[sub_posts['subreddit.name'] == subreddit]
    
    # Reset the index and drop unnecessary columns
    sub_posts_filtered.reset_index(drop=True, inplace=True)
    sub_posts_filtered.drop(columns=['subreddit.id', 'subreddit.name'], inplace=True)
    
    # Save the filtered data to a CSV file
    sub_posts_filtered.to_csv(f'data/subreddits/{subreddit}_posts.csv', index=False)
    
    # Store the length of the current subreddit data
    posts_lengths[subreddit] = len(sub_posts_filtered)
```

```python
# Merge the year on the prerocessed data to filter them
yearly_posts = posts.merge(raw_posts[['id', 'created_utc']], on='id', how='left')

# Convert date code into readable date
yearly_posts["created_utc"] = pd.to_datetime(yearly_posts["created_utc"], unit='s').dt.year

# Initialize an empty dictionary to store the lengths of each year's data
post_lengths = {}

# Loop through the years 2010 to 2022
for year in range(2010, 2023):
    # Filter the data for the current year
    yearly_post_data = yearly_posts[yearly_posts['created_utc'] == year]
    
    # Reset the index
    yearly_post_data.reset_index(drop=True, inplace=True)
    
    # Drop the 'created_utc' column
    yearly_post_data.drop(columns=['created_utc'], inplace=True)
    
    # Save to a CSV file with the year in the filename
    yearly_post_data.to_csv(f'data/yearly_data/posts_{year}.csv', index=False)
    
    # Store the length of the current year's data
    post_lengths[year] = len(yearly_post_data)
```

```python
print("\nLength of post datasets:")
for subreddit in subreddits:
    print(f"{subreddit.capitalize()}: {posts_lengths[subreddit]}")
    

# Print the lengths of datasets for each year
print("\nLength of post post datasets:")
for year in range(2010, 2023):
    print(f"{year}: {post_lengths[year]}")
```
